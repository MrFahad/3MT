```python
# import json
# from googletrans import Translator, LANGUAGES

# # Initialize the Translator
# translator = Translator()

# # Function to translate text in the chunks
# def translate_text(json_data, src_language='id', dest_language='en'):
#     translated_data = json_data.copy()  # Create a copy of the original data

#     # Check if the 'chunks' key exists in the data
#     if 'chunks' not in translated_data:
#         print("Warning: 'chunks' key not found in JSON data.")
#         return translated_data

#     # Iterate through each chunk and translate the 'text'
#     for chunk in translated_data['chunks']:
#         if 'text' in chunk:
#             original_text = chunk['text']
#             try:
#                 # Translate the text
#                 translated_text = translator.translate(original_text, src=src_language, dest=dest_language).text
#                 chunk['text'] = translated_text  # Update the chunk with the translated text
#             except Exception as e:
#                 print(f"Error translating chunk text: {e}")
#         else:
#             print("Warning: 'text' key not found in chunk.")

#     # Translate the main text (outside the chunks)
#     if 'text' in translated_data:
#         try:
#             translated_data['text'] = translator.translate(translated_data['text'], src=src_language, dest=dest_language).text
#         except Exception as e:
#             print(f"Error translating main text: {e}")

#     return translated_data

# # Function to read JSON from file, translate, and save back to a new file
# def process_json_file(input_file, output_file, src_language='id', dest_language='en'):
#     try:
#         # Read the JSON file
#         with open(input_file, 'r', encoding='utf-8') as file:
#             content = file.read()  # Read the file content as a string

#             # Print the content for debugging
#             print("File content:", content)

#             # Load the content as JSON
#             json_data = json.loads(content)

#         # Check if the file content is empty or None
#         if not json_data:
#             raise ValueError("The JSON file is empty or invalid.")

#         # Translate the JSON data
#         translated_json = translate_text(json_data, src_language, dest_language)

#         # Save the translated JSON data to a new file
#         with open(output_file, 'w', encoding='utf-8') as file:
#             json.dump(translated_json, file, indent=4, ensure_ascii=False)

#         print(f"Translation complete. Translated file saved as {output_file}")

#     except FileNotFoundError:
#         print(f"Error: The file '{input_file}' was not found.")
#     except json.JSONDecodeError:
#         print(f"Error: Failed to decode JSON from the file '{input_file}'.")
#     except Exception as e:
#         print(f"An error occurred: {e}")

# # Example usage
# input_file = 'output.json'
# output_file = 'soccer_goal_output.json'

# process_json_file(input_file, output_file, src_language='id', dest_language='en')

import json
from googletrans import Translator

# Initialize the Translator
translator = Translator()

# Function to translate text in the chunks
def translate_text(json_data, dest_language='en'):
    translated_data = json_data.copy()  # Create a copy of the original data

    # Check if the 'chunks' key exists in the data
    if 'chunks' not in translated_data:
        print("Warning: 'chunks' key not found in JSON data.")
        return translated_data

    # Iterate through each chunk and translate the 'text'
    for chunk in translated_data['chunks']:
        if 'text' in chunk:
            original_text = chunk['text']
            try:
                # Detect the source language dynamically
                src_language = translator.detect(original_text).lang
                # Translate the text
                translated_text = translator.translate(original_text, src=src_language, dest=dest_language).text
                chunk['text'] = translated_text  # Update the chunk with the translated text
            except Exception as e:
                print(f"Error translating chunk text: {e}")
        else:
            print("Warning: 'text' key not found in chunk.")

    # Translate the main text (outside the chunks)
    if 'text' in translated_data:
        try:
            src_language = translator.detect(translated_data['text']).lang
            translated_data['text'] = translator.translate(translated_data['text'], src=src_language, dest=dest_language).text
        except Exception as e:
            print(f"Error translating main text: {e}")

    return translated_data

# Function to read JSON from file, translate, and save back to a new file
def process_json_file(input_file, output_file, dest_language='en'):
    try:
        # Read the JSON file
        with open(input_file, 'r', encoding='utf-8') as file:
            content = file.read()  # Read the file content as a string

            # Print the content for debugging
            print("File content:", content)

            # Load the content as JSON
            json_data = json.loads(content)

        # Check if the file content is empty or None
        if not json_data:
            raise ValueError("The JSON file is empty or invalid.")

        # Translate the JSON data
        translated_json = translate_text(json_data, dest_language)

        # Save the translated JSON data to a new file
        with open(output_file, 'w', encoding='utf-8') as file:
            json.dump(translated_json, file, indent=4, ensure_ascii=False)

        print(f"Translation complete. Translated file saved as {output_file}")

    except FileNotFoundError:
        print(f"Error: The file '{input_file}' was not found.")
    except json.JSONDecodeError:
        print(f"Error: Failed to decode JSON from the file '{input_file}'.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
input_file = 'output.json'
output_file = 'soccer_goal_output_v1.json'

process_json_file(input_file, output_file, dest_language='en')
```

```python
import os
import subprocess
from pathlib import Path

def extract_audio(video_path, output_audio_path):
    command = [
        "ffmpeg", "-y", "-i", str(video_path),
        "-vn", "-acodec", "pcm_s16le", "-ar", "16000", "-ac", "1",  # mono WAV 16kHz
        str(output_audio_path)
    ]
    subprocess.run(command, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

def transcribe_with_whisper(audio_path, output_json_path):
    command = [
        "insanely-fast-whisper",
        "--file-name", str(audio_path),
        "--output", str(output_json_path)
    ]
    subprocess.run(command)

def process_videos(input_path):
    input_path = Path(input_path)
    if input_path.is_file():
        video_files = [input_path]
    else:
        video_files = list(input_path.glob("*.mp4")) + list(input_path.glob("*.mkv")) + list(input_path.glob("*.avi"))

    for video in video_files:
        print(f"🔊 Processing: {video.name}")
        audio_wav = video.with_suffix(".wav")
        json_output = video.with_suffix(".json")

        extract_audio(video, audio_wav)
        transcribe_with_whisper(audio_wav, json_output)

        # Optional cleanup
        audio_wav.unlink()

    print("✅ All video files processed.")
```
